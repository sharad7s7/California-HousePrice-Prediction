```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("housing-data.csv")
```

```python
df['income_cat']=pd.cut(df['median_income'], bins=[0,1.5,3,4.5,6,np.inf], labels=[1,2,3,4,5])
```

```python
df
```

```python
import matplotlib.pyplot as plt
df["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(df,df['income_cat']):
    strat_train_data=df.loc[train_index]
    strat_test_data=df.loc[test_index]
```

```python
strat_test_data["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances") 
```

```python
strat_train_data["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances") 
```
